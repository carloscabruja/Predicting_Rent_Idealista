# Transform the raw data to a clean data set

The objective of this notebook is to transform the raw data to a clean data set. Because this will be automated executed it could occur that the data_raw_corrupted file exists so this process will unify the data_raw_corrupted file with the data_raw file and produce the data_clean file.

```python
# Libraries
import pandas as pd
import numpy as np
import ast

# config 
pd.set_option('display.max_columns', None)
```

```python
# Load the data_raw and data_raw_corrupted files
df_raw = pd.read_csv('../data/raw/data_raw.csv')
df_raw
```

```python
df_raw_corrupted = pd.read_csv('../data/raw/data_raw_corrupted.csv')
df_raw_corrupted
```

The corrupted file have less rows so we will compare the propertyCode and append it to have a complete data set.

```python
# Add news propertyCode of data_raw_corrupted to data_raw
df_raw_completed = pd.concat([df_raw, df_raw_corrupted])
df_raw_completed
```

```python
# drop duplicates
df_raw_completed.drop_duplicates(inplace=True)
df_raw_completed
```

So we have a unified data set, but we need to clean it and transform to a data set that can be used to do the analysis.

So the nan in newDevelopmentFinished is not a real NaN it because it not a new development but a finished development.
so we can create a new column named 'is_Finished' and it only be true when newDevelopmentFinished is True and NaN.

```python
df_raw_completed['isFinished'] = df_raw_completed['newDevelopmentFinished'].apply(
    lambda x: False if x == False else True)

df_raw_completed.drop(['newDevelopmentFinished', 'newDevelopment'], axis=1, inplace=True)
df_raw_completed['isFinished'].value_counts()
```

### parkingSpace

```python
df_raw_completed['parkingSpace'].value_counts()
```

```python
# replace all the ' to " in the parkingSpace column
df_raw_completed['parkingSpace'] = df_raw_completed['parkingSpace'].str.replace('\'', '"')

# convert the string to a dictionary
df_raw_completed['parkingSpace'] = df_raw_completed['parkingSpace'].apply(
    lambda x: ast.literal_eval(x) 
    if type(x) == str else x
    )

# get the hasParkingSpace of the dict of the parkingSpace column
df_raw_completed['hasParkingSpace'] = df_raw_completed['parkingSpace'].apply(
    lambda x: x['hasParkingSpace'] if type(x) == dict else False
    )

# get isParkingSpaceIncludedInPrice 
df_raw_completed['isParkingSpaceIncludedInPrice'] = df_raw_completed['parkingSpace'].apply(
    lambda x: x['isParkingSpaceIncludedInPrice'] if type(x) == dict else False
    )

# get the parkingSpacePrice 
df_raw_completed['parkingSpacePrice'] = df_raw_completed['parkingSpace'].apply(
    lambda x: x['parkingSpacePrice'] 
    if type(x) == dict and 'parkingSpacePrice' in x else 0
    )

# drop the parkingSpace column
df_raw_completed.drop(columns=['parkingSpace'], inplace=True)
df_raw_completed.sample(5)
```

### floor

```python
df_raw_completed[df_raw_completed['floor'].isnull()]
```

To unknown's values of floor we will replace them with the mode of the column group by propertyType.

```python
floor_nan = df_raw_completed[df_raw_completed['floor'].isnull()]
floor_nan['propertyType'].value_counts()
```

So we need to know the mode floor of the chalet, flat, country house, etc.

```python
property_type = floor_nan['propertyType'].value_counts().index.to_list()

for i in property_type:
    # get the mode of the propertyType column
    mode = df_raw_completed['floor'][df_raw_completed['propertyType'] == i].mode()[0]
    # get the index of the rows with the propertyType i and the floor is null
    index = df_raw_completed[(df_raw_completed['propertyType'] == i) & (df_raw_completed['floor'].isnull())].index
    # replace the nan with the mode
    df_raw_completed.loc[index, 'floor'] = mode
```

### hasLift

```python
df_raw_completed[df_raw_completed['hasLift'].isnull()]
```

```python
df_raw_completed[df_raw_completed['hasLift'].isnull()]['floor'].value_counts()
```

Because the majority of the values of floor are bj or 1, we will assume that rest of values are False.

```python
index = df_raw_completed[df_raw_completed['hasLift'].isnull()].index
df_raw_completed.loc[index, 'hasLift'] = False
```

### Suspicious types:
* floor
* hasLift

```python
df_raw_completed['floor'].value_counts()
```

```python
df_raw_completed[df_raw_completed['floor'] == 'st']
```

We can input 0 into 'bj' and 'en' because this is the ground floor and mezzanine. And put -1 into 'st' because is a basement.

```python
# where floor is 'bj' or 'en' put 0
df_raw_completed.loc[df_raw_completed['floor'] == 'bj', 'floor'] = 0
df_raw_completed.loc[df_raw_completed['floor'] == 'en', 'floor'] = 0

# where floor is 'st' put 1
df_raw_completed.loc[df_raw_completed['floor'] == 'st', 'floor'] = -1

# set floor as int
df_raw_completed['floor'] = df_raw_completed['floor'].astype(int)
```

```python
df_raw_completed['hasLift'].value_counts()
```

```python
# set hasLift as boolean
df_raw_completed['hasLift'] = df_raw_completed['hasLift'].astype(bool)
```
